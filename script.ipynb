```python
from helper import CreditDataProvider
import pandas as pd

# Switch between 'dummy' and 'live' here:
DATA_MODE = "dummy"
provider = CreditDataProvider(mode=DATA_MODE)
market_data = provider.get_market_snapshot()

# Flattening the term structure for display
display_rows = []
for tenor, spread in market_data['cds_curve'].items():
    display_rows.append({"Asset": f"Intel {tenor}Y CDS", "Value": f"{spread} bps"})
display_rows.append({"Asset": "Bond Price", "Value": market_data['bond']['price']})
display_rows.append({"Asset": "10Y Treasury", "Value": f"{market_data['treasury_curve'].get(10.0, 0)*100:.2f}%"})

df_summary = pd.DataFrame(display_rows)
print(f"--- Step 1: Data Acquisition ({DATA_MODE} mode) ---")
print(df_summary)
```

```python
from helper import HazardRateEngine
# --- Run the Engine ---
engine = HazardRateEngine(market_data)

print("--- Step 2: Bootstrapped Hazard Rates (Lambda) ---")
for t in sorted(engine.hazard_rates.keys()):
    print(f"Tenor {t}Y: {engine.hazard_rates[t]:.4f}")

test_t = 7.0 # Our Intel 2033 Bond Maturity
print(f"\nSurvival Prob to {test_t}Y (Intel 2033 Maturity): {engine.survival_prob(test_t):.2%}")
```
